```python
# Import Libraries
import warnings
import pandas as  pd
```

```python
# Load Dataset
warnings.filterwarnings('ignore')
df = pd.read_csv("Dataset for Data Analytics - Sheet1.csv")
```

```python

df.head()
```

```python
# Checking Rows & Columns
print(f'Number of Rows: {df.shape[0]}')
print(f'Number of Columns: {df.shape[1]}')
```

```python
#Standardize Text Columns Remove extra spaces.
df.columns =df.columns.str.strip()
```

```python
# Checking Duplicate & Null Values
print(f'Duplicate Values are: {df.duplicated().sum()}')
print(f'Null values are: {df.isnull().sum().sum()}')
```

    Checking Null Values

```python
df.isna().sum()
```

```python
df['CouponCode'].unique()
```

```python
df['CouponCode'].value_counts()
```

```python
# NaN simply means, Customer didn't use a coupon.
# Fill Null Values as "No Coupon"

df['CouponCode'] = df['CouponCode'].fillna('No Coupon')
```

```python
print(f'Null values are: {df.isnull().sum().sum()}')
```

    Checking Calculation is right or wrong

```python
incorrect_total = df[df['TotalPrice'] != (df['Quantity'] * df['UnitPrice']).round(2)]

print(len(incorrect_total))
```
