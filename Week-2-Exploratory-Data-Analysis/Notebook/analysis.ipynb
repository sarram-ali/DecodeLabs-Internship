```python
import warnings
import pandas as pd
```

```python
warnings.filterwarnings('ignore')
df = pd.read_csv('cleaned_dataset.csv')
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
summary = pd.DataFrame({
    "Mean": df.mean(numeric_only=True),
    "Median": df.median(numeric_only=True),
    "Min": df.min(numeric_only=True),
    "Max": df.max(numeric_only=True),
    "Std": df.std(numeric_only=True)
})

summary
```

```python
df.describe(include='all')
```

```python
product_revenue = df.groupby('Product')['TotalPrice'].sum().sort_values(ascending=False).reset_index()

product_revenue
```

```python
quantity_sold = df.groupby('Product')['Quantity'].sum().sort_values(ascending=False).reset_index()

quantity_sold
```

```python
Q1 = df["Quantity"].quantile(0.25)
Q3 = df["Quantity"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR

outliers = df[(df["Quantity"]<lower)|(df["Quantity"]>upper)]

print(f'Number of Outliers are: {outliers.shape[0]}')
```

```python
monthly_sales = df.groupby(df['Date'].dt.month)['TotalPrice'].sum().reset_index()

monthly_sales['TotalPrice'] = monthly_sales['TotalPrice'].apply(lambda x: f"${x:,.0f}")

monthly_sales
```

```python
monthly_quantity = df.groupby(df['Date'].dt.month)['Quantity'].sum().reset_index()

monthly_quantity['Quantity'] = monthly_quantity['Quantity'].apply(lambda x: f"{x:,.0f}")

monthly_quantity
```
